```kotlin
%use lets-plot
%use dataframe
import java.math.*
```

```kotlin
val df = DataFrame.read("bench.csv")
val seq = df
    .filter { it["narzędzie"] == "sekwencyjnie" }
```

```kotlin
val plot = letsPlot(df.filter{it["narzędzie"] != "sekwencyjnie" && it["gęstość"] == BigDecimal.valueOf(0.6)}.toMap()) {
    x = "rozmiar macierzy"
    y = "czas"
    color = "narzędzie"
} +
    geomLine(size = 1.0) +
    ggtitle("Czas obliczeń w sekundach")
plot
```

```kotlin
var df = DataFrame.read("PORR/bench.csv")
val seq = df
    .filter { it["narzędzie"] == "sekwencyjnie" }
df = df.add("przyspieszenie") { row ->
    val size = row["rozmiar macierzy"] as Int
    val time1 = row["czas"] as BigDecimal
    
    if (row["narzędzie"] == "sekwencyjnie") {
        1.0
    } else {
           val seqTime = seq.firstOrNull { it["rozmiar macierzy"] == size && it["gęstość"] == row["gęstość"]}!!["czas"] as BigDecimal
        seqTime / time1
    }
}.add("gęstość macierzy") { it["gęstość"].toString() + " " }
df
```

```kotlin
val plot = letsPlot(df.filter{
    it["narzędzie"] != "sekwencyjnie" && 
    it["gęstość"] == BigDecimal.valueOf(0.6)
}.toMap()) {
    x = "rozmiar macierzy"
    y = "przyspieszenie"
    color = "narzędzie"
} +
    geomLine(size = 1.0) +
    ggtitle("Przyspieszenie obliczeń")
plot
```

```kotlin
val plot = letsPlot(df.filter{
    (it["narzędzie"] == "OpenCL" || it["narzędzie"] == "CUDA")
    && it["gęstość"] == BigDecimal.valueOf(0.6)
}.toMap()) {
    x = "rozmiar macierzy"
    y = "przyspieszenie"
    color = "narzędzie"
} +
    geomLine(size = 1.0) +
    ggtitle("GPU")
ggsave(plot, "speed_GPU.png")
```

```kotlin
val plot = letsPlot(df.filter{
    it["narzędzie"] == "MPI" 
}.toMap()) {
    x = "rozmiar macierzy"
    y = "przyspieszenie"
    color = "gęstość macierzy"
} +
    geomLine(size = 1.0) +
    ggtitle("procesy")
plot
```

```kotlin
var df2 = DataFrame.read("cpu.csv")
df2 = df2.add("przyspieszenie") { row ->
    val size = row["rozmiar macierzy"] as Int
    val time1 = row["czas"] as BigDecimal
    
    if (row["narzędzie"] == "sekwencyjnie") {
        1.0
    } else {
           val seqTime = seq.firstOrNull { it["rozmiar macierzy"] == size && it["gęstość"] == row["gęstość"]}!!["czas"] as BigDecimal
        seqTime / time1
    }
}.add("gęstość macierzy") { it["gęstość"].toString() + " " }.add("CPU") { it["cpu"].toString() + "" }
df2
```

```kotlin
val plot = letsPlot(df2.filter{it["cpu"] == 16 && it["gęstość"] == BigDecimal.valueOf(0.6)}.toMap()) {
    x = "rozmiar macierzy"
    y = "przyspieszenie"
    color = "narzędzie"
} +
    geomLine(size = 1.0) +
    ggtitle("Przyspieszenie obliczeń CPU")
plot
```

```kotlin
val plot = letsPlot(df2.filter{it["narzędzie"] == "MPI" && it["gęstość"] == BigDecimal.valueOf(0.6)}.toMap()) {
    x = "rozmiar macierzy"
    y = "przyspieszenie"
    color = "CPU"
} +
    geomLine(size = 1.0) +
    ggtitle("MPI")
plot
```
