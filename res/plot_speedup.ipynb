```kotlin
var df = DataFrame.read("PORR/bench.csv")
val seq = df
    .filter { it["narzÄ™dzie"] == "sekwencyjnie" }
df = df.add("przyspieszenie") { row ->
    val size = row["rozmiar macierzy"] as Int
    val time1 = row["czas"] as BigDecimal
    
    if (row["narzÄ™dzie"] == "sekwencyjnie") {
        1.0
    } else {
           val seqTime = seq.firstOrNull { it["rozmiar macierzy"] == size && it["gÄ™stoÅ›Ä‡"] == row["gÄ™stoÅ›Ä‡"]}!!["czas"] as BigDecimal
        seqTime / time1
    }
}.add("gÄ™stoÅ›Ä‡ macierzy") { it["gÄ™stoÅ›Ä‡"].toString() + " " }
df
```

```kotlin
val plot = letsPlot(df.filter{
    it["narzÄ™dzie"] != "sekwencyjnie" && 
    it["gÄ™stoÅ›Ä‡"] == BigDecimal.valueOf(0.6)
}.toMap()) {
    x = "rozmiar macierzy"
    y = "przyspieszenie"
    color = "narzÄ™dzie"
} +
    geomLine(size = 1.0) +
    ggtitle("Przyspieszenie obliczeÅ„")
plot
```

```kotlin
val plot = letsPlot(df.filter{
    (it["narzÄ™dzie"] == "OpenCL" || it["narzÄ™dzie"] == "CUDA")
    && it["gÄ™stoÅ›Ä‡"] == BigDecimal.valueOf(0.6)
}.toMap()) {
    x = "rozmiar macierzy"
    y = "przyspieszenie"
    color = "narzÄ™dzie"
} +
    geomLine(size = 1.0) +
    ggtitle("GPU")
ggsave(plot, "speed_GPU.png")
```
